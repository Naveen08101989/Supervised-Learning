```python
#Create independent variable list  - by removing outcome column

df =df.drop(['outcome'],axis=1)
print (df)
```

```python
#Print the number of missing values and Column Data Types 

dtype_null = pd.concat([df.isnull().sum(),df.dtypes], axis = 1)
dtype_null.columns= ['Null_Count','Datatype']
dtype_null
```

```python
#Columns with missing values and data type – Object 

category_cols= dtype_null[(dtype_null.Null_Count >0)& (dtype_null.Datatype=='object')]
category_cols
```

```python
#Get Column names as List 

category_variables = list(category_cols.index)
category_variables
```

```python
#Missing value update for Category Column 

from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy="most_frequent")
df[category_variables]= imp.fit_transform(df[category_variables])
df
```

```python
#Check missing value count 

dtype_null = pd.concat([df.isnull().sum(),df.dtypes], axis = 1)
dtype_null.columns= ['Null_Count','Datatype']
dtype_null
```

```python
#Save Result to Excel to Validate 

df.to_excel ('CS32.xlsx')
```

```python
#Float columns with missing record count 

float_cols= dtype_null[(dtype_null.Null_Count >0)]
float_cols
```

```python
#Float column names as list  

float_variables = list(float_cols.index)
float_variables
```

```python
#Missing value update for Float columns with mean 

from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy="mean")
df[float_variables]= imp.fit_transform(df[float_variables])
df.to_excel ('CS32.xlsx')
df
```

```python
#Validate the data  

df[float_variables]
```

```python
#Display independent table list- Data Set 

df
```

```python
#Get all the columns list data type – Object 


allcategory_cols= dtype_null[dtype_null.Datatype=='object']
allcategory_cols= list(allcategory_cols.index)
allcategory_cols
```

```python
#Implement one hot encoding  


df1= df
for i in allcategory_cols:
    df1=pd.get_dummies(df1, columns=[i], drop_first=True)
df1.to_excel ('CS32.xlsx')
df1
```

```python
#Target Column Unique values   

target.unique()
```

```python
#Label encoding for Target column 

from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
target= label.fit_transform(target)
print (target)   
```

```python
#Train Test Split 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1, target, test_size=0.2, random_state=1)


print (X_train.shape)

print (X_test.shape)
```

```python
#Decision Tree  


from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
#Prediction x_test 

ydt_predict = dt.predict(X_test)
ydt_predict
```

```python
# Print Accuracy  

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ydt_predict,y_test)
print(accuracy)
```

```python
#5.	Fit a random forest classifier and observe the accuracy.

#Random Forest 

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)  
```

```python
#Prediction x_test 

yrf_predict = rf.predict(X_test)
yrf_predict
```

```python
#Print Accuracy  
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(yrf_predict,y_test)
print(accuracy)
```
